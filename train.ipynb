```python
import cv2
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
```

```python
# 모델 평가 함수
def evaluate_model(model, test_data, test_labels):
    test_loss, test_acc = model.evaluate(test_data, test_labels)
    print(f"테스트 정확도: {test_acc}")

# 모델 예측 함수
def predict_with_model(model, new_data):
    predictions = model.predict(new_data)
    return predictions
```

```python
def grayscaling(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    return img

def color(img_path):
    img = cv2.imread(img_path)
    return img

def normalizing(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    img = cv2.normalize(img, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
    return img

def Eqalize(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    img = cv2.equalizeHist(img)
    return img

def GaussianBlu(img_path):
    img = cv2.imread(img_path)
    kernel_size = (5, 5)
    img = cv2.GaussianBlur(img, kernel_size, 0)



```

```python
# NumPy의 난수 시드 고정
np.random.seed(0)

# TensorFlow의 난수 시드 고정 (2.x 버전)
tf.random.set_seed(0)


# train 이미지 폴더 경로
data_dir = "C:/Temp/CV/2023-uou-hd-dt-2-cv/train/"
classes = ['0', '1']

data = []
labels = []

for class_label in classes:
    class_path = os.path.join(data_dir, class_label)
    for img_file in os.listdir(class_path)[:100]:
        img_path = os.path.join(class_path, img_file)
        Eqalize(img_path)
        img = cv2.resize(img, (28, 28))
        data.append(img)
        labels.append(classes.index(class_label))

data = np.array(data)
labels = np.array(labels)


# CNN 모델 생성
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

# 모델 컴파일
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 모델 훈련
model.fit(data, labels, epochs=5)

# 임시 라벨링
predict_label = [1,0,0,0,0,0,1,1,0,0]
predict_label = np.array(predict_label)

# 모델 평가
test_loss, test_acc = model.evaluate(data, labels)
print(f"테스트 정확도: {test_acc}")

predict_folder = "C:/Temp/CV/2023-uou-hd-dt-2-cv/test"
predict_data = os.listdir(predict_folder)[:10]
predict = []
for i in range(len(predict_data)):
    predict_data_path = predict_folder + "/" + predict_data[i]
    predict_img = GaussianBlu(predict_data_path)
    predict_img = cv2.resize(predict_img, (28, 28))
    predict.append(predict_img)

predict = np.array(predict)
# 모델 예측
predictions = model.predict(predict)


evaluate_model(model, predict, predict_label)
predict_folder = "C:/Temp/CV/2023-uou-hd-dt-2-cv/test"
eval_data = os.listdir(predict_folder)[11:20]
eval = []  # 예측할 이미지 데이터를 로드
for i in range(len(eval_data)):
    eval_data_path = predict_folder + "/" + eval_data[i]
    eval_img = Eqalize(eval_data_path)
    eval_img = cv2.resize(eval_img, (28, 28))
    eval.append(eval_img)
    
eval = np.array(eval)
```
